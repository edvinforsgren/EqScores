```python
import pandas as pd
import utilities as ut
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df_all_prec_recall = pd.read_csv("./data/all_prec_recall_values.csv")
```

```python
for pert in df_all_prec_recall.Perturbation.unique().tolist():
    fig = ut.create_grouped_scatter_plot(df_all_prec_recall.query(f'Perturbation == "{pert}"').sort_values('Time', ascending=False).sort_values('Modality'))
    fig.savefig(f'./figures/knn_prec_recall_{pert}_loocv_no_q2.pdf', format='pdf', bbox_inches='tight')
    fig.show()
```
